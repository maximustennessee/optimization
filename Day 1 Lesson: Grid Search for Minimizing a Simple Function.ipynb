Define the Function
Create a function in Python to represent ùëì(ùë•):

```python
def f(x):
    
    return (x-3) ** 2 + 5
```

Create a range of x values

```python
import numpy as np

x_values = np.arange(0, 6.1, 0.1)  # Generates values from 0 to 6 (inclusive) in steps of 0.1
```

Evaluate the Function and Find the Minimum

```python
best_x = None
best_value = float('inf')

for x in x_values:
    current_value = f(x)
    if current_value < best_value:
        best_x = x
        best_value = current_value

print("Minimum x = {} & minimum f(x) = {}".format(best_x, best_value))
```

Plot the Function

```python
import matplotlib.pyplot as plt

# Evaluate f(x) for all x values for plotting
y_values = [f(x) for x in x_values]

plt.figure(figsize=(8, 4))
plt.plot(x_values, y_values, label=f'f(x) = {best_value:.2f}')
plt.scatter(best_x, best_value, color='red', label=f'Minimum at x={best_x:.2f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Minimizing f(x) using Grid Search')
plt.legend()
plt.grid(True)
plt.show()
```
